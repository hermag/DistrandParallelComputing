```python
# ─────────────────────────────────────────────────────────────────────────────
# IMPORTS
# (assumed to be at the top of your file)
# ─────────────────────────────────────────────────────────────────────────────

import os          # Provides access to operating system features
                   # We use it to get the Process ID (PID)

import threading   # Provides threading utilities
                   # We use it to get the Thread ID (TID) of the current thread

import time        # Provides time-related functions
                   # We use time.sleep() to simulate a slow task

import datetime    # Provides date and time utilities
                   # We use it to measure how long the whole program takes


# ─────────────────────────────────────────────────────────────────────────────
# FUNCTION DEFINITION
# ─────────────────────────────────────────────────────────────────────────────

def print_something(sleep_time, something_to_print):
    """
    Simulates a task that takes some time to complete.

    Parameters:
        sleep_time (int)         : how many seconds the task will take
        something_to_print (str) : a label to identify which call this is
    """

    # Announce that the function has started and show which call this is.
    print(f"This is print_something function with argument {something_to_print}")

    # Simulate doing real work (e.g. reading a file, calling an API, querying a DB)
    # by pausing execution for `sleep_time` seconds.
    #
    # ⚠️  KEY POINT FOR SEQUENTIAL EXECUTION:
    # The program is completely BLOCKED here until sleep finishes.
    # Nothing else can happen while we wait — this is what "sequential" means.
    time.sleep(sleep_time)

    # The function returns here — only AFTER the sleep is fully done.


# ─────────────────────────────────────────────────────────────────────────────
# STATUS DICTIONARY
# ─────────────────────────────────────────────────────────────────────────────

# A dictionary used to track and display the current state of the program.
# We use a dictionary (instead of a plain string) so it can be easily updated
# from anywhere in the code — useful later when we move to multithreading.
status = {"text": "🟡 Starting..."}

# Print the initial status BEFORE the main block runs.
# This line executes immediately when the script is loaded.
print(status["text"])


# ─────────────────────────────────────────────────────────────────────────────
# MAIN BLOCK
# ─────────────────────────────────────────────────────────────────────────────

# This condition is True only when the script is run directly (not imported).
# It is a Python best practice to put your "entry point" logic here.
if __name__ == "__main__":

    # Get the Process ID (PID) — a unique number the OS assigns to this program.
    # Every running program has its own PID.
    pid = os.getpid()

    # Get the Thread ID (TID) — a unique number identifying the current thread.
    # Since we are not using multithreading yet, there is only ONE thread: the main thread.
    # Compare this with the multithreading version where each thread has a different TID!
    tid = threading.get_ident()

    # Update the status to "Running" and include PID and TID for observability.
    status["text"] = f"🔵 Running... | PID: {pid} | Thread ID: {tid}"
    print(status["text"])

    # Record the exact moment the main logic starts.
    # We will compare this with end_time later to measure total duration.
    start_time = datetime.datetime.now()

    print("This is the main function")

    # ── FIRST CALL (SEQUENTIAL) ───────────────────────────────────────────────
    #
    # Set up the arguments for the first call.
    sleep_time = 5
    something_to_print = "First Call"

    # Call print_something() DIRECTLY — no threads involved.
    # The program will STOP HERE and wait the full 5 seconds before moving on.
    # This is sequential execution: one thing at a time, in order.
    print_something(sleep_time, something_to_print)

    # ── SECOND CALL (SEQUENTIAL) ──────────────────────────────────────────────
    #
    # The program only reaches this line AFTER the first call fully finishes.
    sleep_time = 10
    something_to_print = "Second Call"

    # Again, the program blocks here for 10 seconds before continuing.
    # Total blocking time so far: 5 + 10 = 15 seconds.
    print_something(sleep_time, something_to_print)

    # ── MEASURE TOTAL TIME ────────────────────────────────────────────────────

    # Record the time now that both calls have finished.
    end_time = datetime.datetime.now()

    # Subtract start from end to get the total elapsed time.
    # You should see ~15 seconds here: 5s (first call) + 10s (second call).
    # This is the "cost" of sequential execution — tasks add up one after another.
    print(f"It took {end_time - start_time} seconds")

    # Update the status to "Done" and print it.
    status["text"] = f"✅ Done! | PID: {pid} | Thread ID: {tid}"
    print(status["text"])

    # ── SUMMARY ───────────────────────────────────────────────────────────────
    #
    # In this sequential version:
    #   - There is only ONE thread (the main thread) — TID never changes
    #   - Tasks run ONE AT A TIME, in the exact order they are written
    #   - Total time = sum of all individual task times (5 + 10 = 15s)
    #
    # In the multithreading version you will see:
    #   - MULTIPLE threads with different TIDs running at the same time
    #   - The main thread finishing in milliseconds
    #   - Total time ≈ the longest single task (10s), not the sum
```

## Multithreading without waiting for the spawned threads

```python
# ─────────────────────────────────────────────────────────────────────────────
# IMPORTS
# ─────────────────────────────────────────────────────────────────────────────

import threading   # Allows us to create and manage threads
                   # A thread is a lightweight unit of execution within a process

import os          # Provides a way to interact with the operating system
                   # We use it here to get the Process ID (PID)

import time        # Provides time-related functions
                   # We use time.sleep() to simulate a task that takes time

import datetime    # Provides classes for working with dates and times
                   # We use it to measure how long the execution takes


# ─────────────────────────────────────────────────────────────────────────────
# FUNCTION DEFINITION
# ─────────────────────────────────────────────────────────────────────────────

def print_something(sleep_time, something_to_print):
    """
    Simulates a task that takes some time to complete.
    
    Parameters:
        sleep_time (int)         : how many seconds the task will "work" (sleep)
        something_to_print (str) : a label so we can identify which call this is
    """

    # os.getpid() returns the Process ID (PID) of the current process.
    # All threads share the same PID because they live inside the same process.
    pid = os.getpid()

    # threading.get_ident() returns the Thread ID (TID) of the thread
    # that is currently executing this line.
    # Each thread has a UNIQUE TID — this is how we tell threads apart.
    tid = threading.get_ident()

    # Print a START message so we can see exactly when this thread began working.
    # Including PID and TID helps students observe:
    #   - same PID  → same process
    #   - different TID → different thread
    print(f"[PID: {pid} | TID: {tid}] START - print_something with argument '{something_to_print}'")

    # Simulate doing some work by sleeping for `sleep_time` seconds.
    # In a real program this could be a database query, file read, API call, etc.
    # Crucially: while THIS thread is sleeping, OTHER threads can keep running!
    time.sleep(sleep_time)

    # Print an END message so we can see exactly when this thread finished.
    print(f"[PID: {pid} | TID: {tid}] END - print_something with argument '{something_to_print}'")


# ─────────────────────────────────────────────────────────────────────────────
# MAIN BLOCK
# ─────────────────────────────────────────────────────────────────────────────

# This condition is True when the script is run directly (not imported as a module).
# It is a Python best practice to put your "entry point" code here.
if __name__ == "__main__":

    # Get the PID and TID of the MAIN thread so we can compare them
    # with the threads we are about to create.
    pid = os.getpid()
    tid = threading.get_ident()

    print(f"🔵 Running... | PID: {pid} | Thread ID: {tid}")

    # Record the exact time we start, so we can measure total elapsed time later.
    start_time = datetime.datetime.now()

    print("This is the main function")

    # ── CREATE THREADS ────────────────────────────────────────────────────────
    #
    # threading.Thread() creates a new thread but does NOT start it yet.
    #
    #   target = the function the thread will run
    #   args   = the arguments passed to that function, as a tuple
    #
    # t1 will call: print_something(5,  "First Call")
    # t2 will call: print_something(10, "Second Call")
    #
    # Note: we do NOT use parentheses when passing the function:
    #   ✅  target=print_something
    #   ❌  target=print_something()   ← this would call it immediately!

    t1 = threading.Thread(target=print_something, args=(5,  "First Call"))
    t2 = threading.Thread(target=print_something, args=(10, "Second Call"))

    # ── START THREADS ─────────────────────────────────────────────────────────
    #
    # .start() tells the OS to begin executing the thread.
    # After t1.start(), t1 runs CONCURRENTLY with the main thread.
    # The main thread does NOT wait for t1 to finish before moving on!

    t1.start()  # t1 begins running print_something(5,  "First Call")  in background
    t2.start()  # t2 begins running print_something(10, "Second Call") in background

    # ── MAIN THREAD CONTINUES IMMEDIATELY ────────────────────────────────────
    #
    # The main thread reaches this point almost instantly (microseconds),
    # while t1 and t2 are still sleeping in the background.
    # This is the key demonstration of NON-SEQUENTIAL (concurrent) execution!
    #
    # In the sequential version the total time would be 5 + 10 = 15 seconds.
    # Here the main thread is already done while the others are still running.

    end_time = datetime.datetime.now()
    print(f"Main thread done in {end_time - start_time} seconds")
    print(f"✅ Done! | PID: {pid} | Thread ID: {tid}")
```

## Multithreading with join - to wait for all spawned threads

```python
# ─────────────────────────────────────────────────────────────────────────────
# IMPORTS
# ─────────────────────────────────────────────────────────────────────────────

import threading   # Allows us to create and manage threads
                   # A thread is a lightweight unit of execution within a process

import os          # Provides a way to interact with the operating system
                   # We use it here to get the Process ID (PID)

import time        # Provides time-related functions
                   # We use time.sleep() to simulate a task that takes time

import datetime    # Provides classes for working with dates and times
                   # We use it to measure how long the execution takes


# ─────────────────────────────────────────────────────────────────────────────
# FUNCTION DEFINITION
# ─────────────────────────────────────────────────────────────────────────────

def print_something(sleep_time, something_to_print):
    """
    Simulates a task that takes some time to complete.
    
    Parameters:
        sleep_time (int)         : how many seconds the task will "work" (sleep)
        something_to_print (str) : a label so we can identify which call this is
    """

    # os.getpid() returns the Process ID (PID) of the current process.
    # All threads share the same PID because they live inside the same process.
    pid = os.getpid()

    # threading.get_ident() returns the Thread ID (TID) of the thread
    # that is currently executing this line.
    # Each thread has a UNIQUE TID — this is how we tell threads apart.
    tid = threading.get_ident()

    # Print a START message so we can see exactly when this thread began working.
    # Including PID and TID helps students observe:
    #   - same PID  → same process
    #   - different TID → different thread
    print(f"[PID: {pid} | TID: {tid}] START - print_something with argument '{something_to_print}'")

    # Simulate doing some work by sleeping for `sleep_time` seconds.
    # In a real program this could be a database query, file read, API call, etc.
    # Crucially: while THIS thread is sleeping, OTHER threads can keep running!
    time.sleep(sleep_time)

    # Print an END message so we can see exactly when this thread finished.
    print(f"[PID: {pid} | TID: {tid}] END - print_something with argument '{something_to_print}'")


# ─────────────────────────────────────────────────────────────────────────────
# MAIN BLOCK
# ─────────────────────────────────────────────────────────────────────────────

if __name__ == "__main__":

    # Get the PID and TID of the MAIN thread so we can compare them
    # with the threads we are about to create.
    pid = os.getpid()
    tid = threading.get_ident()

    print(f"🔵 Running... | PID: {pid} | Thread ID: {tid}")

    # Record the exact time we start, so we can measure total elapsed time later.
    start_time = datetime.datetime.now()

    print("This is the main function")

    # ── CREATE THREADS ────────────────────────────────────────────────────────
    #
    # threading.Thread() creates a new thread but does NOT start it yet.
    #
    #   target = the function the thread will run
    #   args   = the arguments passed to that function, as a tuple
    #
    # Note: we do NOT use parentheses when passing the function:
    #   ✅  target=print_something
    #   ❌  target=print_something()   ← this would call it immediately!

    t1 = threading.Thread(target=print_something, args=(5,  "First Call"))
    t2 = threading.Thread(target=print_something, args=(10, "Second Call"))

    # ── START THREADS ─────────────────────────────────────────────────────────
    #
    # .start() tells the OS to begin executing the thread.
    # Both threads are now running CONCURRENTLY in the background.
    # The main thread does NOT wait for either of them to finish.

    t1.start()  # t1 begins running print_something(5,  "First Call")  in background
    t2.start()  # t2 begins running print_something(10, "Second Call") in background

    # ── JOIN THREADS ──────────────────────────────────────────────────────────
    #
    # .join() makes the MAIN THREAD pause and WAIT at this line until
    # the specified thread has fully finished.
    #
    # Without join():
    #   - The main thread would rush past and print "Done" almost instantly,
    #     while t1 and t2 are still running in the background.
    #
    # With join():
    #   - t1.join() → main thread waits here until t1 finishes (~5s)
    #   - t2.join() → main thread then waits here until t2 finishes (~10s)
    #
    # ⚠️  IMPORTANT: Even with join(), t1 and t2 still ran CONCURRENTLY!
    #   - t1 and t2 were both sleeping AT THE SAME TIME after .start()
    #   - join() only controls when the MAIN THREAD is allowed to continue
    #   - Total time will still be ~10s (not 15s), because both threads
    #     overlapped — join() does not make execution sequential!

    t1.join()   # Main thread blocks here until t1 is completely done
    t2.join()   # Main thread blocks here until t2 is completely done

    # ── MAIN THREAD RESUMES ───────────────────────────────────────────────────
    #
    # We only reach this line AFTER both t1 and t2 have finished.
    # This guarantees that all work is complete before we print the final summary.
    # This is useful when subsequent code depends on the results of the threads.

    end_time = datetime.datetime.now()

    # Because both threads ran concurrently, total time ≈ 10s (the longest task),
    # NOT 15s (5 + 10) as it would be in the sequential version.
    print(f"All threads done in {end_time - start_time} seconds")
    print(f"✅ Done! | PID: {pid} | Thread ID: {tid}")

    # ── SUMMARY ───────────────────────────────────────────────────────────────
    #
    # Without join():  main thread prints "Done" in milliseconds,
    #                  t1 and t2 finish later on their own.
    #
    # With join():     main thread WAITS for t1 and t2,
    #                  then prints "Done" only after ALL work is complete.
    #
    # Either way, t1 and t2 ran concurrently — join() only affects
    # WHEN the main thread is allowed to move forward, not how the
    # child threads execute relative to each other.
```
