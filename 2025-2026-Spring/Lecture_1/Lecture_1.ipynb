## Sequential Code Example

```python
from IPython.display import clear_output
import time, threading, os
import datetime
```

```python
def print_something(sleep_time, something_to_print):
    print(f"This is print_something function with argument {something_to_print}")
    time.sleep(sleep_time)
```

```python
status = {"text": "🟡 Starting..."}
print(status["text"])
if __name__ == "__main__":
    pid = os.getpid()
    tid = threading.get_ident()
    status["text"] = f"🔵 Running... | PID: {pid} | Thread ID: {tid}"
    print(status["text"])
    start_time = datetime.datetime.now()
    print("This is the main function")
    sleep_time = 5
    something_to_print = "First Call"
    print_something(sleep_time, something_to_print)
    sleep_time = 10
    something_to_print = "Second Call"
    print_something(sleep_time, something_to_print)
    end_time = datetime.datetime.now()
    print(f"It took {end_time - start_time} seconds")
    status["text"] = f"✅ Done! | PID: {pid} | Thread ID: {tid}"
    print(status["text"])
```
