Python interpreter is Good enough to avoid trivial race conditions, however - here is the actual case

```python
import threading
import time

counter = 0
the_range = 100_000
number_of_threads = 5
def unsafe_increment():
    global counter
    for _ in range(the_range):
        temp = counter           # Read shared value
        time.sleep(0.0000001)    # Force thread switching
        temp += 1                # modify local copy
        counter = temp           # write the value back

threads = []

for _ in range(number_of_threads):
    t = threading.Thread(target=unsafe_increment)
    threads.append(t)
    t.start()

for t in threads:
    t.join()

print("Final counter value:", counter)
print(f"Expected value: {number_of_threads*the_range}")
```
