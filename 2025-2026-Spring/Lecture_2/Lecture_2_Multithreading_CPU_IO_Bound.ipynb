Example of two classes which, later, will be used for the multithreading

```python
class Hello:
    def do(self):
        for i in range(5):
            print("Hello ",i+1)

class Hi:
    def do(self):
        for i in range(5):
            print("Hi",i+1)
```

```python
t1 = Hello()
t2 = Hi()
```

```python
t1.do()
t2.do()
```

This is not the goal, goal is to run all this simultaniously, at least to try to run everything at the same time.
Examples of such behaviours are everywhere, in our phone, ipad, or computer oparating systems and applications. For instance in case of games, lots of events are happening at the same time, e.g. changing the score, moving the objects, playing the music etc.

```python
#import the necessary package for multithreading
from threading import Thread
```

```python
#inherit classes from Thread, and change do methods to run - run is the key method in Thread super class and in this way
#we are overriding the run method
class Hello(Thread):
    def run(self):
        for i in range(5):
            print("Hello ",i+1)

class Hi(Thread):
    def run(self):
        for i in range(5):
            print("Hi",i+1)
```

```python
t1 = Hello()
t2 = Hi()
```

```python
t1.run()
t2.run()
```

The above code, still calls the class methods, and not starting the threads! in order to start the threads, we need to use the keyword start

```python
#Let's add sleep part to feel the multithreading part
from time import sleep
class Hello(Thread):
    def run(self):
        for i in range(5):
            print("Hello ",i+1)
            sleep(0.3)

class Hi(Thread):
    def run(self):
        for i in range(5):
            print("Hi",i+1)
            sleep(0.3)
            
t1 = Hello()
t2 = Hi()
t1.start()
t2.start()
```

```python
#Just to be sure that sleep does not change the main behaviour, lets use run once again
t1.run()
t2.run()
```

A bit cleaner implementation of threads using classes

```python
import time
import threading
class DownloadThread(threading.Thread):
    """Thread that downloads a URL and stores result."""

    def __init__(self, url):
        super().__init__(daemon=True) # always call super().__init__()
        self.url = url
        self.result = None            # shared via instance attribute

    def run(self):
        print(f"Fetching {self.url}")
        #simulate network request
        time.sleep(1)
        self.result = f"<html from {self.url}>"

urls = ["https://example.com","https://python.org"]
threads = [DownloadThread(u) for u in urls]

for t in threads: t.start()           # starts the threads
for t in threads: t.join()            # guarantees that all threads will be executed and only after that this cell's execution will exit
```

Let's make one more example of downloading the files, we will pass arguments to the functions, through the threads module

```python
from time import sleep, time

def download(file_name):
    print("Downloading file...", file_name)
    sleep(0.5)
    print("Download complete")

files = ['video.mp4','image.png','data.csv']
start = time()
for f in files:
    download(f)
end = time()
print(f"serial time {end - start:.2f} seconds")
print("Bye")
```

The same, as above but implementation with threads, we are facing here advantage of I/O bound task

```python
from threading import Thread
from time import sleep, time

threads = []

def download(file_name):
    print("Downloading file...", file_name)
    sleep(0.5)
    print("Download complete")

files = ['video.mp4','image.png','data.csv']

for f in files:
    t = Thread(target=download, args=(f,))
    threads.append(t)

start = time()
for t in threads:
    t.start()

for t in threads:
    t.join()
    
end = time()
print(f"threaded time {end - start:.2f} seconds")
print("Bye")
```

Let's see how multithreading will behave when we are dealing with CPU bound task

```python
from threading import Thread
from time import sleep, time

def calculate(n1,n2):
    sum = 0
    for n in range(n1,n2):
        sum+=n*n
```

```python
num = 50_000_000
start = time()
calculate(0,num)
end = time()
print(f"threaded time {end - start:.2f} seconds")
```

Let's implement the same with threads, and split the calculations in two equal parts

```python
num = 50_000_000
mid = num // 2

t1 = Thread(target = calculate, args = (0,mid))
t2 = Thread(target = calculate, args = (mid, num))
start = time()
t1.start()
t2.start()
t1.join()
t2.join()
end = time()
print(f"threaded time {end - start:.2f} seconds")
```
